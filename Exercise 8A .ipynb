2 Write a function to implement the **linear search**. 

**def linearSearch(A, n, key):**
```
# Search the n-element integer array A for a match with key.
# Return the index position of the key item if found, or -1 otherwise.
```

Complete the main function to test the linear search by determining the index position of a key in a list.
```
<Sample output 1 >                      <Sample output 2>
Number of integers in the array: 4      Number of integers in the array: 3  
Enter integer 1: 20                     Enter integer 1: 5 
Enter integer 2: 30                     Enter integer 2: 8
Enter integer 3: 10                     Enter integer 3: 2
Enter integer 4: 50
                                        [5, 8, 2]
[20, 30, 10, 50]                        Enter a key: 15 
Enter a key: 10                         15 is not found!
10 is found at index positon 2
```


````python
def linearSearch(A, n, key):

    # Search the n-element integer array A for a match with key.
    # Return the index position of the key item if found, or -1 otherwise.
    i = 0
    while i <= n-1:
        if A[i] == key:
            return i
        else:
            i += 1
    else:
        return -1

def main():
    numList = []     # initialize an empty list
    n = int(input("Number of integers in the array: "))

    for i in range(n):
        num = int(input(f"Enter integer {i+1}:"))
        numList.append(num)
    
    print ()
    print (numList)

    key = int(input("Enter a key: "))

    result = linearSearch(numList, n, key)
    if result == -1:
        print('Key is not found!')
    else:
        print(f'Key is found in index {result}')

main()
````

5 Write a function to implement the **binary search**

**def binarySearch(A, n, key):**
```
# Search the n-element sorted integer array A for a match with key.
# Return the index position of the key item if found, or -1 otherwise.
```
Complete the main function to test the binary search by determining the index position of a key in a list.
```
<Sample output 1 >                      <Sample output 2>
Number of integers in the array: 4      Number of integers in the array: 3 
Enter integers in ascending order       Enter integers in ascending order
Enter integer 1: 20                     Enter integer 1: 5 
Enter integer 2: 30                     Enter integer 2: 8
Enter integer 3: 40                     Enter integer 3: 9
Enter integer 4: 50
                                        [5, 8, 9]
[20, 30, 40, 50]                        Enter a key: 15 
Enter a key: 30                         15 is not found!
30 is found at index positon 1
```

````python
def binarySearch(A, n, key):
    # Search the n-element sorted integer array A for a match with key.
    # Return the index position of the key item if found, or -1 otherwise.
    low = 0
    high = n-1
    while low<=high:
        middle = (low+high) // 2
        if A[middle] == key:
            return middle
        elif A[middle] < key:
            low = middle + 1
        else:
            high = middle - 1
    return -1
#-----------------------------------------------------------------
def main():
    numList =[]     # initialize an empty list
    n = int(input("Number of integers in the array: "))

    # prompt for and enter a list of integers
    print ("Enter integers in ascending order.")    
    for i in range(n):
        num = int(input(f"Enter integer {i+1}:"))
        numList.append(num)
    
    print ()
    print (numList)

    key = int(input("Enter a key: "))
    result = binarySearch(numList, n, key)
    if result != -1:
        print(f'Index is {result}')
    else:
        print('Not found')
main()
````

6  Write a function to implement the **binary search**

**def binSearch(A, low, high, key):**
```
# Search a sorted array A, A[low] . . . . A[high], for a match with key using binary search. 
# Return the index of the matching array element or -1 if a match does not occur
```
Complete the main function to test the binary search by determining the index position of a key in a list.
```
<Sample output 1 >                      <Sample output 2>
Number of integers in the array: 4      Number of integers in the array: 3 
Enter integers in ascending order       Enter integers in ascending order
Enter integer 1: 20                     Enter integer 1: 5 
Enter integer 2: 30                     Enter integer 2: 8
Enter integer 3: 40                     Enter integer 3: 9
Enter integer 4: 50
                                        [5, 8, 9]
[20, 30, 40, 50]                        Enter a key: 15 
Enter a key: 30                         15 is not found!
30 is found at index positon 1
```

````python
def binSearch(A, low, high, key):
    # Search a sorted array A, A[low] . .. A[high], for a match with key using the binary search
    # Return the index of the matching array element or -1 if a match does not occur
    while low<=high:
        middle = (low+high) // 2
        if A[middle] == key:
            return middle
        elif A[middle] < key:
            low = middle + 1
        else:
            high = middle - 1
    return -1

#-----------------------------------------------------------------
def main():
    numList =[]     # initialize an empty list
    n = int(input("Number of integers in the array: "))

    # prompt for and enter a list of integers
    print ("Enter integers in ascending order.")    
    for i in range(n):
        num = int(input(f"Enter integer {i+1}:"))
        numList.append(num)
    
    print ()
    print (numList)

    key = int(input("Enter a key: "))
    low = int(input("Enter low: "))
    high = int(input("Enter high: "))
    result = binSearch(numList, low, high, key)
    if result != -1:
        print(f'Index is {result}')
    else:
        print(f'Not found between index {low} and {high}')
        
main()
````
