1.  Write a function to implement the 2-way merge.

**def TwoWayMerge(A, B, C):**
```
# merge two sorted arrays A and B into one single sorted array C
```

````python
def TwoWayMerge(A, B, C):
    m = len(A)
    n = len(B)
    i, j, k = 0, 0, 0  # i is pointer of A, j is pointer of B, k is pointer of C
    while i < m and j < n:
        if A[i] < B[j]:
            C[k] = A[i]
            i += 1
            k += 1
        else:
            C[k] = B[j]
            j += 1
            k += 1
    if i < m:
        C[k:] = A[i:]
    elif j < n:
        C[k:] = B[j:]
    return C
    
def main():

    # Driver code to test above 
    A = [2, 7, 10, 12, 16, 19]   # sorted
    B = [4, 6, 8, 11, 13]        # sorted
 
    print ("Given sorted array A:") 
    print(*A)
    print ("Given sorted array B:") 
    print(*B)

    C = [0] * (len(A)+len(B))  
    TwoWayMerge(A, B, C)

    print ("Merged array C:") 
    print(*C)

main()

````
