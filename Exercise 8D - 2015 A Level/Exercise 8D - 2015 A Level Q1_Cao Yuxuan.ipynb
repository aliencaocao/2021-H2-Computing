```python
def bubbleSort(data):
    noSwaps = False
    count = 0
    while not noSwaps:
        noSwaps = True
        for i in range(len(data)-1):
            count += 1
            if data[i] > data[i+1]:
                noSwaps = False
                temp = data[i]
                data[i] = data[i+1]
                data[i+1] = temp
    print(f'Number of comparsions: {count}')
    return data


def insertSort(data):
    n = len(data)
    count = 0
    for i in range(1, n):
        target = data[i]
        j = i
        while j > 0 and data[j-1] > target:
            count += 1
            data[j] = data[j-1]
            j -= 1
        data[j] = target
    print(f'Number of comparsions: {count}')
    return data


def main():
    stop = False
    while not stop:
        print('1. Read file data\n2. Bubble sort\n3. Insertion Sort\n4. End')
        choice = int(input('Enter a choice: '))
        if choice == 1:
            with open('ADMISSIONS-DATA.txt', 'r') as f:
                num = [int(i) for i in f.read().splitlines()]
        elif choice == 2:
            try:
                print(*bubbleSort(num))
            except NameError:
                print('Please run read file first!')

        elif choice == 3:
            try:
                print(*insertSort(num))
            except NameError:
                print('Please run read file first!')
        elif choice == 4:
            stop = True

main()
```
