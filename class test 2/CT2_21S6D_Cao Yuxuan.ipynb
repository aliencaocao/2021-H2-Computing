```python
# Cao Yuxuan 21S6D
# Task 1.1
class Student:
    def __init__(self, name, height):
        self.full_name = name
        self.height = float(height)

    def getName(self):
        return self.full_name

    def getHeight(self):
        return self.height

    def setName(self, name):
        self.full_name = name

    def setHeight(self, height):
        self.height = float(height)

    def compareHeight(self, student):
        assert isinstance(student, Student)
        if self.height == student.height:
            return 0
        elif self.height > student.height:
            return -1
        else:
            return 1

    def printInfo(self, prnt=True):
        string = f'{self.full_name}[{self.height}]'
        if prnt:
            print(string)
        return string

# Task 1.2
def task1_2(filename):
    students = []
    with open(filename, 'r') as f:
        lines = f.read().splitlines()
        for line in lines:
            students.append(Student(*line.split(',')))
    return students

result = task1_2('STUDENT.txt')
for stu in result:
    stu.printInfo()
```

```python
# Task 1.3
def task1_3(students):
    def swap(a, b):
        temp = a
        a = b
        b = temp
        return a, b
    def partition(A, start, end, key):
        # check if key is an attribute of every item in A
        assert all([hasattr(a, key) for a in A]), 'Key is not an attribute of every item in A!'
        i = start + 1  # skip the first element as using it as pivot
        # Array A below is a list of class instances, not just height. During comparsion, we only compare the specified key (which is height in this case), but we rearrange/swap around the entire instance.
        pivot = A[start]
        for j in range(start+1, end+1):
            if getattr(A[j], key) < getattr(pivot, key):  # could have used student.getHeight() here but I wanted to be more generic
                A[i], A[j] = swap(A[i], A[j])
                i += 1
        A[start], A[i-1] = A[i-1], A[start]
        return i-1

    def quicksort(A, start, end, key):  # key is the attribute we want to sort by
        if start < end:
            pivot_pos = partition(A, start, end, key)
            quicksort(A, start, pivot_pos-1, key)
            quicksort(A, pivot_pos+1, end, key)
        return A

    sorted_students = quicksort(students, 0, len(students)-1, key='height')
    return sorted_students

result = task1_3(task1_2('STUDENT.txt'))
for stu in result:
    stu.printInfo()
```

```python
# Task 2.1
class Student:
    def __init__(self, name, height):
        self.full_name = name
        self.height = float(height)

    def compareHeight(self, student):
        assert isinstance(student, Student)
        if self.height == student.height:
            return 0
        elif self.height > student.height:
            return -1
        else:
            return 1

    def printInfo(self, prnt=True):
        string = f'{self.full_name}[{self.height}]'
        if prnt:
            print(string)
        return string


class StudentNode(Student):
    def __init__(self, name, height):
        super().__init__(name, height)
        self.nxt = None

    def getNext(self):
        return self.nxt

    def setNext(self, nxt):
        assert isinstance(nxt, StudentNode)
        self.nxt = nxt


# Task 2.2
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert(self, node):
        # Question asks for insertation at the head, so it's a prepend
        assert isinstance(node, StudentNode)
        if not self.head:
            self.head = node
            self.tail = node
        else:
            node.setNext(self.head)
            self.head = node

    def delete(self, name):
        probe = self.head
        if probe.getName() == name:
            self.head = self.head.getNext()
            return True
        found = False
        while probe and not found:
            if probe.getNext().getName() == name:
                probe.setNext(probe.getNext().getNext())
                found = True
        if found:
            return True
        else:
            return False

    def to_String(self):
        probe = self.head
        return_str_list = []
        while probe:
            return_str_list.append(probe.printInfo(prnt=False))
            probe = probe.getNext()
        return ', '.join(return_str_list)


# Task 2.3
def task2_3(filename):
    studentsLinkedList = LinkedList()
    with open(filename, 'r') as f:
        lines = f.read().splitlines()
        for line in lines:
            studentsLinkedList.insert(StudentNode(*line.split(',')))
    return studentsLinkedList

result = task2_3('STUDENT.txt')
print(f'[{result.to_String()}]')


# Task 2.4
class Queue(LinkedList):
    def enqueue(self, node):
        # Enqueue is at the end so we cannot use insert which is prepend
        assert isinstance(node, StudentNode)
        if not self.head:
            self.head = node
            self.tail = node
        else:
            self.tail.setNext(node)
            self.tail = node

    def dequeue(self):
        if not self.head:
            return None
        else:
            popped_head = self.head
            self.head = self.head.getNext()
            return popped_head

student_queue = Queue()
with open('STUDENT.txt', 'r') as f:
    lines = f.read().splitlines()
    for line in lines:
        student_queue.enqueue(StudentNode(*line.split(',')))

print(student_queue.to_String())
for i in range(3):
    student_queue.dequeue().printInfo()
```
