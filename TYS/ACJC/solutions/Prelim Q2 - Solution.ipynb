```python
class Array():
    def __init__(self,n):
        self.array = []
        for i in range(n-1):
            self.array.append([None, i+1, None])
        self.array.append([None, None, None])
        self.freepointer = 0
        
    def set_freepointer(self,x):
        self.freepointer = x
    
    def set_node_data(self,x,data):
        self.array[x][0] = data
    
    def set_node_lpointer(self,x,y):
        self.array[x][1] = y
    
    def set_node_rpointer(self,x,y):
        self.array[x][2] = y

    def get_freepointer(self):
        return self.freepointer

    def get_node_data(self,x):
        return self.array[x][0]
    
    def get_node_lpointer(self,x):
        return self.array[x][1]
    
    def get_node_rpointer(self,x):
        return self.array[x][2]   

class BST():
    def __init__(self, array):
        self.array = array
        self.rootpointer = None
    
    def add_node(self,data):
        
        # Freelist is empty
        if self.array.get_freepointer() == None:
            print("No available nodes")
        
        else:
            temp = self.array.get_freepointer()
            
            # Move freelist pointer to next item of freelist
            self.array.set_freepointer(self.array.get_node_lpointer(temp))
            
            # Tree is initially empty
            if self.rootpointer == None:
                self.rootpointer = temp
                self.array.set_node_data(self.rootpointer,data)
                
            else:
                #look for correct place to insert the new node
                current = self.rootpointer
                prev = self.rootpointer
                while current != None:
                    prev = current
                    if self.array.get_node_data(current) < data:
                        current = self.array.get_node_rpointer(current)
                    else:
                        current = self.array.get_node_lpointer(current)
                current = temp
                self.array.set_node_data(current,data)
                if self.array.get_node_data(prev) < data:
                    self.array.set_node_rpointer(prev,temp)
                else:
                    self.array.set_node_lpointer(prev,temp)
            self.array.set_node_lpointer(temp,None)

    def search_tree(self,data):
        node = self.rootpointer
        while node != None:
            if self.array.get_node_data(node) < data:
                node = self.array.get_node_rpointer(node)
            elif self.array.get_node_data(node) > data:
                node = self.array.get_node_lpointer(node)
            else:
                return node
        print("Not found")
        return None
            
    def in_order_traversal(self):
        def in_order_traversal_helper(node):
            if node == None:
                return []
            else:
                result = []
                result.extend(in_order_traversal_helper(self.array.get_node_lpointer(node)))
                result.append(self.array.get_node_data(node))
                result.extend(in_order_traversal_helper(self.array.get_node_rpointer(node)))
                return result
        return in_order_traversal_helper(self.rootpointer)
      
```

```python
        

myArray = Array(12)
myBST = BST(myArray)
yourBST = BST(myArray)

myBST.add_node(50)
print(myArray.array)
print(myBST.rootpointer)
myBST.add_node(25)
print(myArray.array)

yourBST.add_node(50)
yourBST.add_node(75)

myBST.add_node(75)
yourBST.add_node(25)
print(myArray.array)

print(myBST.in_order_traversal())
print(yourBST.in_order_traversal())
```
