```python
# Task 1.1
def BinarySearch(lst, item):
    low = 0
    high = len(lst)-1
    found = False
    while low <= high and not found:
        mid = (low+high) // 2
        temp = lst[mid]
        if item == temp:
            found=True
        elif item < temp:
            high = mid - 1
        else:
            low = mid + 1
    if found:
        return mid
    elif lst[high] == item:
        return high
    else:
        return -1

with open('1000WORDS.TXT', 'r', encoding='utf-8') as f:
    dataset = f.read().splitlines()

def search(item):
    result = BinarySearch(dataset, item)
    if result == -1:
        print('Not found')
    else:
        print(f'Found at {result}')

search('WORD')
search('WORDA')
search('TRADE')
search('YOURSELF')
```

```python

# Task 1.2

def BinarySearch(lst, item):
    low = 0
    high = len(lst)
    results = []
    start = -1
    while low <= high and start==-1:
        mid = (low+high) // 2
        temp = lst[mid][:len(item)]
        if item == temp:
            start = mid
        elif item < temp:
            high = mid - 1
        else:
            low = mid + 1
    if start == -1:
        return -1
    else:
        for i in lst[start:]:
            if i[:len(item)] == item:
                results += [i]
            else:
                return results

with open('1000WORDS.TXT', 'r') as f:
    dataset = f.read().splitlines()

def search(item):
    result = BinarySearch(dataset, item)
    if result == -1:
        print('Not found')
    else:
        print(f'Found: {result}')

search('TR')
search('RE')
```
