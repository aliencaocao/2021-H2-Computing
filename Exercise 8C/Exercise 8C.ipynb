````python
def bubbleSort_2(A, n):   #improved bubble sort
# sort the n-element integer array A (A[0], . . . ,A[n-1]) in ascending order
    stop = False
    run = 0
    while not stop:
        run += 1
        Aold = A.copy()
        for i in range(0, n-1):
           if A[i] > A[i+1]:
               swap(A, i, i+1)
        if Aold == A:
            stop = True
    print(run)

def  main():
    A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
    n = len(A)          # number of elements in the list
    
    print ("Original List: ")
    printList(A, n)

    bubbleSort_2 (A, n)

    print ("Sorted List: ")
    printList (A, n)


main()    
````

````python
def bubbleSort_3(A, n):
# sort the n-element integer array A (A[0], . . . . ,A[n-1]) in ascending order
    last = n-1
    run = 0
    while last > 0:
        lastChange = 0
        run+=1
        for j in range(0, last):
            if A[j] > A[j+1]:
                swap(A, j, j+1)
                lastChange = j
        last = lastChange
    print(run)
def main():
    A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
    n = len(A)          # number of elements in the list
    
    print ("Original List: ")
    printList(A, n)

    bubbleSort_3(A, n)

    print ("Sorted List: ")
    printList (A, n)


main()
````
