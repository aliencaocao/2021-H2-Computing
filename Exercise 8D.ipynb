2 Write a function to implement the **insertion sort**.

**def insertionSort(A, n):**
```
# sort the n-element integer array A (A[0], . . . ,A[n-1]) in ascending order
```

````python
def insertionSort(A, n):
    for i in range(1, n):
        target = A[i]
        print(f'Got new target {target}')
        j = i
        while j > 0 and A[j-1] > target:
            print(f'Prev no is larger than target, {A[j-1]} > {target}')
            print(f'Set the current term to be prev term ({A[j]} -> {A[j-1]})')
            A[j] = A[j-1]
            j -= 1
        print(f'Prev term is smaller than target ({A[j-1]} < {target}). Current term {A[j]} is set to {target}')
        A[j] = target
        print(A)
    return A


def  main():
    A = [38, 58, 13, 15, 51, 27, 10, 19, 12, 86, 49, 67, 84, 60, 25]
    n = len(A)          # number of elements in the list
    
    print ("Original List: ")
    printList(A, n)

    A = insertionSort (A, n)

    print ("Sorted List: ")
    printList (A, n)

main()    
````
